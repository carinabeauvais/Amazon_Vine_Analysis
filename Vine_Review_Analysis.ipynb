```python
import pandas as pd
```

```python
vine_df = pd.read_csv("/vine_table.csv")
vine_df
```

```python
total_votes = vine_df[(vine_df["total_votes"] >= 20)]
total_votes
```

```python
# get percent of helpful votes
total_votes["helpful_votes_percentage"] = (total_votes["helpful_votes"]) / (total_votes["total_votes"])* 100
total_votes
```

```python
# percents over 50%
over50_votes = total_votes[(total_votes["helpful_votes_percentage"] >= 50)]
over50_votes
```

```python
paid_votes = over50_votes[(over50_votes["vine"] == "Y")]
paid_votes
```

```python
unpaid_votes = over50_votes[(over50_votes["vine"] == "N")]
unpaid_votes
```

```python
total_paid = paid_votes["review_id"].count()
total_paid
```

```python
total_unpaid = unpaid_votes["review_id"].count()
total_unpaid
```

```python
fivestar_paid = paid_votes[(paid_votes["star_rating"] == 5)].count()
fivestar_paid
```

```python
fivestar_unpaid = unpaid_votes[(unpaid_votes["star_rating"] == 5)].count()
fivestar_unpaid
```

```python
percent_paid_fivestar = (fivestar_paid) / (total_paid) *100
percent_paid_fivestar
```

```python
percent_unpaid_fivestar = (fivestar_unpaid) / (total_unpaid) *100
percent_unpaid_fivestar
```
